### Agrupamento de datas
Diversas colunas em ambos sets são marcações de tempo, que podem ajudar a inferir a duração dum tratamento.

Antes de convertê-las ao formato `datetime`, é necessário testar algumas hipóteses de preenchimento incorreto de datas.
Como primeiro passo para tratá-las, devemos convertê-las ao formato `datetime`.

```python
def typo_list(yr):
    return [yr[1] + yr[0] + yr[2:],
            yr[:2] + yr[3] + yr[2],
            yr[0] + yr[2] + yr[1] + yr[3],
            yr[1] + yr[0] + yr[3] + yr[2]]

year_typos = [typo_list(str(year)) for year in range(2014, 2020)]
year_typos
```

```python
year_typos = [item for sublist in year_typos for item in sublist]
```

```python
typo_collection = pd.DataFrame()

for yt in year_typos:
    pd.concat([typo_collection, df_mama_quim[['dt_soli']][df_mama_quim.dt_soli.astype(str).str.find(yt) == 1]], ignore_index=True)

typo_collection
```

Esgotada a possibilidade de que haja erros de digitação por posicionamento de caracteres, devemos analisar datas inválidas, no critério:

- Data acima do dia `20/10/2019`, dia do recebimento dos dados da competição. Claramente datas acima desta apontam para um procedimento "futuro" inexistente
- Data em má formatação (e.g. com caracteres que não sejam dígitos ou barra `/` de separação)

#### Tratamento de datas em má formatação 

```python
find_non_numeric = lambda x: not str(x).replace('e+', '').replace('.', '').isnumeric()
```

```python
df_mama_rad[['dt_auto']][df_mama_rad.dt_auto.fillna(0).apply(find_non_numeric)]
```

```python
df_mama_quim[['dt_auto']][df_mama_quim.dt_auto.fillna(0).apply(find_non_numeric)]
```

É perceptível a invalidez destes dados - o campo referente ao ano em cada uma destas datas não foi preenchido, permanecendo então seu placeholder, `aaa` ou `aaaa`. Podemos seguramente retirá-los.

```python
df_mama_quim.loc[df_mama_quim.dt_auto.fillna(0).apply(find_non_numeric), 'dt_auto'] = np.nan
```

#### Tratamento de datas futuras & passadas

Eliminaremos aqui datas que se situem acima do considerado possível (data de recebimento dos dados), ou abaixo do considerado factível (meados do início do século passado). Para isto, definimos uma função auxiliar que trata apenas do primeiro caso - datas futuras.

```python
def find_future_dates(col, maximum=20191020):
    return df_mama_quim[col].fillna(0).astype(str).str.replace('/', '').fillna(0).astype(float).astype(int) > maximum
```

```python
df_mama_quim[['dt_soli']][find_future_dates('dt_soli')].sample(15, random_state=rst)
```

```python
df_mama_quim[['dt_auto']][find_future_dates('dt_auto')].sample(15, random_state=rst)
```

Valores formatados como `2.04703e+07` também são datas futuras, pois se traduzem em `20470500.0`. Eliminaremos portanto tais valores das colunas afetadas.

```python
df_mama_quim.loc[find_future_dates('dt_soli'), 'dt_soli'] = np.nan
df_mama_quim.loc[find_future_dates('dt_auto'), 'dt_auto'] = np.nan
```

Analisando então datas passadas, tomamos como parâmetro datas em `dt_soli` que ocorram antes da década de 1950, longínquos 70 anos antes do ano desta análise.

```python
df_mama_quim[['dt_soli']][df_mama_quim.dt_soli < 19500000].sample(10, random_state=rst)
```

```python
df_mama_quim.loc[df_mama_quim.dt_soli < 19500000, 'dt_soli'] = np.nan
```

#### Conversão de datas

```python
dtcols = [c for c in df_mama_quim.columns if c.startswith('dt_')]
df_mama_quim[dtcols].head(10)
```

```python
df_mama_quim[dtcols].dtypes
```

As duas primeiras colunas - mês de processamento e atendimento ao paciente - formatam-se em ano-mês, portanto terão conversão diferenciada

```python
ymth_dt = dtcols[0:2]
to_year_month = lambda df: df.apply(pd.to_datetime, format='%Y%M')
```

```python
df_mama_quim[ymth_dt] = df_mama_quim[ymth_dt].apply(to_year_month)
df_mama_rad[ymth_dt] = df_mama_rad[ymth_dt].apply(to_year_month)

df_mama_quim[ymth_dt].head(5)
```

```python
ymd_dt = dtcols[2:-1]

# Retirada de casas decimais e conversão para datetime
str_to_date = lambda date: np.nan if pd.isnull(date) or len(date) != 8 else pd.to_datetime(date, format='%Y%m%d')

float_to_ymd = lambda df: df.fillna(0).astype(int).astype(str).replace('0', np.nan).apply(str_to_date)
```

```python
df_mama_quim[ymd_dt] = df_mama_quim[ymd_dt].apply(float_to_ymd)
df_mama_rad[ymd_dt] = df_mama_rad[ymd_dt].apply(float_to_ymd)

df_mama_quim[ymd_dt].head(5)
```

### Conversão de códigos inteiros

O número de APAC anterior, `num_apac_ant`, deve ser tratada de modo a eliminar a ocorrência de múltiplos zeros - o que nos indica que não houve APAC anterior - e por eficiência de armazenamento, ser tipada de acordo.

```python
obj_to_int = lambda s: s.astype(str).str.extract('(\d+)', expand=False).replace('^0*$', '0').fillna(0).astype(float).astype(int)
```

```python
col = 'num_apac_ant'
df_mama_quim[col] = obj_to_int(df_mama_quim[col])
```

A mesma lógica de formatação aplica-se às colunas de código CID:

```python
df_mama_quim['cid_caus_assoc'].sample(10, random_state=rst)
```

```python
for c in [c for c in df_mama_quim.columns if c.startswith('cid_')]:
    df_mama_quim[c] = obj_to_int(df_mama_quim[c])
```

```python
df_mama_quim['cid_caus_assoc'].sample(10, random_state=rst)
```

Por fim, também em similaridade com os formatos supracitados, é observado no Código CNES dos estabelecimentos solicitantes (`cod_soli_cnes`) a mesma necessidade de processamento:

```python
df_mama_quim.cod_soli_cnes.sample(10, random_state=rst)
```

```python
df_mama_quim['cod_soli_cnes'] = obj_to_int(df_mama_quim.cod_soli_cnes)
df_mama_quim['nac_paciente'] = obj_to_int(df_mama_quim.nac_paciente)
```
